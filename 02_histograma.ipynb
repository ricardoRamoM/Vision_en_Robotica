# 02 — Histogramas en OpenCV (Color y Escala de Grises)

Generar Histogramas


```python
# ✅ Importar librerías necesarias
import cv2
import matplotlib.pyplot as plt

# ✅ Cargar  la imagen 
img_bgr = cv2.imread(r"C:\Users\20808\Documents\Repositorios\Vision_en_Robotica\imagenes\lenna.png") # Cambia la ruta según tu estructura de carpetas

# Verificar si la imagen se cargó
if img_bgr is None:
    print("⚠️ No se pudo cargar la imagen. Verifica la ruta.")
else:
    print("✅ Imagen cargada correctamente con tamaño:", img_bgr.shape)
```

## Imagen en color (RGB) y sus histogramas por canal




```python
if img_bgr is not None:
    # Convertir a RGB
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

    # Calcular histogramas
    hist_r = cv2.calcHist([img_rgb], [0], None, [256], [0, 256])
    hist_g = cv2.calcHist([img_rgb], [1], None, [256], [0, 256])
    hist_b = cv2.calcHist([img_rgb], [2], None, [256], [0, 256])

    # Figura combinada
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # Imagen RGB
    axes[0].imshow(img_rgb)
    axes[0].set_title("Imagen en RGB")
    axes[0].axis("off")

    # Histogramas
    axes[1].plot(hist_r, color='r', label='Rojo')
    axes[1].plot(hist_g, color='g', label='Verde')
    axes[1].plot(hist_b, color='b', label='Azul')
    axes[1].set_title("Histogramas por canal (RGB)")
    axes[1].set_xlabel("Intensidad (0-255)")
    axes[1].set_ylabel("Número de píxeles")
    axes[1].legend()

    plt.tight_layout()
    plt.show()
```

**Interpretación del histograma RGB:**  
- Los tres canales cubren gran parte del rango de intensidades, lo cual indica que la imagen tiene **buen contraste**.  
- Si un canal está más desplazado hacia intensidades altas, significa que ese color está más presente en la imagen (por ejemplo, valores altos en R → tonos cálidos).  
- Los picos pronunciados corresponden a áreas grandes con colores similares, como fondos uniformes o piel.  


## Imagen en escala de grises y su histograma


```python
if img_bgr is not None:
    img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

    # Calcular histograma
    hist_gray = cv2.calcHist([img_gray], [0], None, [256], [0, 256])

    # Figura combinada
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # Imagen en gris
    axes[0].imshow(img_gray, cmap="gray")
    axes[0].set_title("Imagen en escala de grises")
    axes[0].axis("off")

    # Histograma gris
    axes[1].plot(hist_gray, color='black')
    axes[1].set_title("Histograma (Gray)")
    axes[1].set_xlabel("Intensidad (0-255)")
    axes[1].set_ylabel("Número de píxeles")

    plt.tight_layout()
    plt.show()
```

**Interpretación del histograma en gris:**  
- Si el histograma está concentrado a la izquierda (0–50), la imagen es **oscura**.  
- Si está concentrado a la derecha (200–255), la imagen es **clara**.  
- Un histograma muy estrecho indica **bajo contraste** (poca diferencia entre claros y oscuros).  
- Un histograma extendido en todo el rango indica **alto contraste**, con zonas oscuras y claras bien diferenciadas.
