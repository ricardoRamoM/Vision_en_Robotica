Este notebook muestra cÃ³mo:

1. Leer una imagen desde archivo con **OpenCV** (`cv2.imread`).
2. Mostrarla usando **Matplotlib**.
3. Convertir de BGR (formato interno de OpenCV) a RGB para visualizar correctamente los colores.
4. Guardar una imagen "transformada" en la carpeta original

```python

# âœ… Importar librerÃ­as necesarias
import cv2
import matplotlib.pyplot as plt

# âœ… Leer la imagen (ruta relativa, por ejemplo desde carpeta imagenes)
## Nota: asegÃºrate de tener una imagen llamada "lenna.png" o cualquier otra en imagenes/
img_bgr = cv2.imread(r"C:\Users\rrm21\Documents\Vision_en_Robotica\imagenes\lenna.png")# Cambia la ruta segÃºn tu estructura de carpetas

# Verificar si la imagen se cargÃ³
if img_bgr is None:
    print("âš ï¸ No se pudo cargar la imagen. Verifica la ruta.")
else:
    print("âœ… Imagen cargada correctamente con tamaÃ±o:", img_bgr.shape)

plt.imshow(img_bgr)
plt.title("Mostrada como BGR")
plt.axis("off")
plt.show()

#print(img_bgr) #arreglo completo
px_bgr=img_bgr[100,100]
print("BGR:",px_bgr)  # Imprime el valor del pÃ­xel en la posiciÃ³n (100, 100)

```

```python

# âœ… Mostrar la imagen con Matplotlib
# OpenCV carga en formato BGR, debemos convertir a RGB para verla correctamente
if img_bgr is not None:
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    plt.title("Imagen cargada con OpenCV")
    plt.axis("off")   # ocultar ejes
    plt.show()

# âœ… Imprimir la matriz NumPy de la imagen
px_rgb=img_rgb[100,100]
print("RGB:",px_rgb)  # Imprime el valor del pÃ­xel en la posiciÃ³n (100, 100)
```

```python

# âœ… Convertir a escala de grises
if img_bgr is not None:
    img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
    plt.imshow(img_gray, cmap="gray")
    plt.title("Imagen en escala de grises")
    plt.axis("off")
    plt.show()

#print(img_gray)
px_gray=img_gray[100,100]
print("Gray:",px_gray)  # Imprime el valor del pÃ­xel en la posiciÃ³n (100, 100)
```

```python

# âœ… Guardar la imagen procesada
if img_bgr is not None:
    cv2.imwrite(r"C:\Users\rrm21\Documents\Vision_en_Robotica\imagenes\lenna_gray.png", img_gray)
    print("âœ… Imagen en escala de grises guardada en imagenes")
```
